```python
%matplotlib inline
%precision 16

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
filepath = "./post_process/"
files = os.listdir(filepath)
```

```python
# Show the domain of certain file by giving all the files and the order of the file
def show_domain(files, file_num):
    iter = 0
    for file in files:
        if not os.path.isdir(file):
            filename = filepath + file
            data = pd.read_csv(filename,sep="\s+",header=1)
            if(iter == file_num+1):
                fig = plt.figure(figsize=(16,8), facecolor='white')
                plt.scatter(data.Position_x, data.Position_y, alpha=0.6)
        iter +=1
        
```

```python
show_domain(files,3)
```
